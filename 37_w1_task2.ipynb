## Subtask 1 & 2

### Linearly Separable and Non-Linear Seperable Data


```python
import numpy as np
np.random.seed(42)

# Linearly separable: Gaussian blobs (class1 top-right, class2 bottom-left)
class1_linear = np.random.randn(100, 2) * 0.8 + [2.5, 1.5]
class2_linear = np.random.randn(100, 2) * 0.7 + [-1.5, -2.0]
linear_data = np.vstack([class1_linear, class2_linear])
linear_labels = np.hstack([np.zeros(100), np.ones(100)])

# Non-linearly separable: rings with variable radius (inner ~1, outer ~2.5)
theta = np.linspace(0, 2*np.pi, 100)
r1 = 1 + 0.2 * np.random.rand(100)
circle1 = np.c_[r1 * np.cos(theta), r1 * np.sin(theta)]
r2 = 2.5 + 0.15 * np.random.rand(100)
circle2 = np.c_[r2 * np.cos(theta), r2 * np.sin(theta)]
nonlinear_data = np.vstack([circle1, circle2])
nonlinear_labels = np.hstack([np.zeros(100), np.ones(100)])

print("Linear shape:", linear_data.shape)
print("Non-linear shape:", nonlinear_data.shape)
```

## Subtask 3
### Plot Datasets

Creation of figure with subplots, having the following features - appropriate labels, legends, and titles ('Linearly Separable' and 'Non-Linearly Separable'), DPI of 150 and `bbox_inches='tight'`.




```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Linear plot
ax1.scatter(linear_data[linear_labels==0, 0], linear_data[linear_labels==0, 1],
            c='blue', label='Class 0', alpha=0.7)
ax1.scatter(linear_data[linear_labels==1, 0], linear_data[linear_labels==1, 1],
            c='red', label='Class 1', alpha=0.7)
ax1.set_title('Linearly Separable Dataset')
ax1.legend(); ax1.grid(True, alpha=0.3)

# Non-linear plot
ax2.scatter(nonlinear_data[nonlinear_labels==0, 0], nonlinear_data[nonlinear_labels==0, 1],
            c='blue', label='Class 0', alpha=0.7)
ax2.scatter(nonlinear_data[nonlinear_labels==1, 0], nonlinear_data[nonlinear_labels==1, 1],
            c='red', label='Class 1', alpha=0.7)
ax2.set_title('Non-Linearly Separable Dataset')
ax2.legend(); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('37-w1-data-plot.png', dpi=150, bbox_inches='tight')
plt.show()

```

### Analysis of Data Characteristics

### Geometric Differences

*   **Linearly Separable Data (class1 and class2):**
    The `class1` and `class2` datasets are generated from Gaussian distributions with distinct mean values. This results in two clusters of points that are clearly separated in space. Geometrically, it means that a single straight line (or a hyperplane in higher dimensions) can be drawn to perfectly divide these two classes. This is evident as `class1` is centered around [2, 2] and `class2` around [-2, -2], creating a clear gap between them.

*   **Non-Linearly Separable Data (circle1 and circle2):**
    The `circle1` and `circle2` datasets consist of two concentric circles. `circle1` forms an inner ring, while `circle2` forms an outer ring. In this configuration, no single straight line can separate the inner circle from the outer circle. Any straight line drawn through the data would inevitably cut through one or both circles, failing to isolate one from the other. Therefore, a curved boundary is necessary to achieve separation.

### Suitable Machine Learning Models

*   **For Linearly Separable Data (e.g., class1 and class2):**
    For datasets where classes can be separated by a straight line, simpler linear models are highly effective and computationally efficient.
    *   **Logistic Regression:** A statistical model that uses a logistic function to model a binary dependent variable.
    *   **Linear Support Vector Machine (Linear SVM):** A model that finds the hyperplane that best separates the two classes with the maximum margin.
    These models are generally preferred due to their interpretability and lower computational cost for linearly separable problems.

*   **For Non-Linearly Separable Data (e.g., circle1 and circle2):**
    For datasets where a curved boundary is required for separation, more complex models capable of learning non-linear relationships are necessary.
    *   **Kernel Support Vector Machine (Kernel SVM):** Uses a kernel trick to transform the data into a higher-dimensional space where it might become linearly separable, effectively finding a non-linear decision boundary in the original space.
    *   **Decision Trees/Random Forests:** Tree-based models can create complex, non-linear decision boundaries by recursively partitioning the feature space.
    *   **Neural Networks:** Deep learning models with multiple layers and non-linear activation functions are inherently capable of learning highly complex and non-linear patterns in data, making them suitable for intricate decision boundaries.
